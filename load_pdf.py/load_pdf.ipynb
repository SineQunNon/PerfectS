## OCR 사용해서 표 추출을 해보자

PaddleOCR을 사용한 표 추출

```python
from paddleocr import PaddleOCR
import fitz #PyMuPDF

# PDF 문서를 이미지로 변환
def pdf_to_images(pdf_path):
    pdf_docs = fitz.open(pdf_path)
    images = []

    for page_num in range(len(pdf_docs)):
        page = pdf_docs.load_page(page_num)
        pix = page.get_pixmap()
        img = pix.tobytes("png")
        images.append(img)
    return images

# PaddleOCR 초기화
ocr = PaddleOCR(use_angle_cls=True, lang='korean')

# PDF에서 이미지 추출
pdf_path = '/Users/sinequanon/Desktop/RAG/Capstone/data/test1.pdf'
images = pdf_to_images(pdf_path)

for i, img in enumerate(images):
    result = ocr.ocr(img)
    print(f"Page {i+1}")
    for line in result:
        print(line)
```

```python
from paddleocr import PaddleOCR
import fitz  # PyMuPDF
from PIL import Image
import io
import os

# PDF 문서를 이미지로 변환
def pdf_to_images(pdf_path):
    pdf_docs = fitz.open(pdf_path)
    images = []

    for page_num in range(len(pdf_docs)):
        page = pdf_docs.load_page(page_num)
        pix = page.get_pixmap()
        img = pix.tobytes("png")
        images.append(img)
    return images

# PNG 파일로 저장
def save_images_as_png(images, output_dir):
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    for i, img_bytes in enumerate(images):
        img = Image.open(io.BytesIO(img_bytes))
        img.save(os.path.join(output_dir, f"page_{i+1}.png"))


pdf_path = '/Users/sinequanon/Desktop/RAG/Capstone/data/test_one.pdf'
images = pdf_to_images(pdf_path)

# 이미지 파일로 저장
#output_dir = '/Users/sinequanon/Desktop/RAG/Capstone/data/images'
#save_images_as_png(images, output_dir)
```

```python
from paddleocr import PaddleOCR

# 텍스트에서 표를 식별하는 함수 (단순한 예시)
def extract_tables_from_ocr_results(ocr_results):
    tables = []
    for result in ocr_results:
        table = []
        for line in result:
            text = line[1][0]
            if '|' in text or '\t' in text:  # 텍스트에서 표 식별 (단순한 조건)
                table.append(text)
        if table:
            tables.append(table)
    return tables

ocr = PaddleOCR(use_angle_cls=True, lang='korean')

# 각 이미지에 대해 OCR 수행
ocr_results = []
for i, img in enumerate(images):
    result = ocr.ocr(img)
    ocr_results.append(result)

# OCR 결과에서 표 추출
tables = extract_tables_from_ocr_results(ocr_results)

# 추출된 표 출력
for page_num, table in enumerate(tables):
    print(f"Page {page_num+1}")
    for row in table:
        print(row)
```
